### 제2절. 막대그래프와 히스토그램

#### 1. 막대그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

wine_load = load_wine()
wine = pd.DataFrame(wine_load.data, columns = wine_load.feature_names)
wine['class'] = wine_load.target
wine['class'] = wine['class'].map({0: 'class_0', 1: 'class_1', 2: 'class_3'})
```

```python
# 도수분포표
wine_type = wine['class'].value_counts()
wine_type
```

```python
plt.bar(wine_type.index, wine_type.values, 
        width=0.8, bottom=None,
        align='center')
# width: 막대의 너비
# bottom: 데이터를 넣고자 하는 막대의 바닥 부분의 y축 위치ㄴ
```

```python
plt.barh(wine_type.index, wine_type.values, 
         height=0.8, left=None, align='edge')
# align: 막대의 x축 좌표에서 정렬을 수행
```

#### 2. 히스토그램

```python
wine
```

```python
plt.title('Wine alcohol histogram')
plt.hist('alcohol', bins=8, range=(11, 15), color='purple', data=wine)
plt.show()
```

### 제3절. 상자 그림(Box plot)

```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['class'] = load_iris().target
iris['class'] = iris['class'].map({0:'Setosa', 1:'Versicolor', 2:'Virginia'})
```

```python
iris
```

```python
iris
```

```python
plt.boxplot(iris.drop(columns='class'), 
            labels = iris.columns[:4])
plt.show()
```

```python
plt.boxplot(iris['sepal width (cm)'], whis=1.0)
plt.title('sepal width (cm)')
plt.show()
```

```python
iris[['sepal width (cm)', 'class']].boxplot(by='class')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='class', y='sepal width (cm)', data=iris, 
            hue='class')
plt.show()
```

### 제4절. 산점도

```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['class'] = load_iris().target
iris['class'] = iris['class'].map({0:'Setosa', 1:'Versicolor', 2:'Virginica'})
```

```python
# 산점도 그리기
plt.title('iris scatter')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(x=iris['sepal length (cm)'], 
            y=iris['sepal width (cm)'], 
            alpha=0.5)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', 
                data=iris, hue='class', style='class')
plt.show()
```

### 제5절. 선그래프

#### 1. 수평선/수직선 그래프

```python
import matplotlib.pyplot as plt

plt.hlines(-6, -10, 10, color='grey')
plt.hlines(-2, -10, 10, color='green')
plt.hlines(2, -10, 10, color='orange')
plt.hlines(6, -10, 10, color='red')
plt.vlines(-6, -10, 10, linestyles='solid')
plt.vlines(-2, -10, 10, linestyles='dashed')
plt.vlines(2, -10, 10, linestyles='dashdot')
plt.vlines(6, -10, 10, linestyles='dotted')
```

#### 2. 함수식 그래프

```python
def linear_func(x):
    return 2 * x + 1

X = iris['sepal length (cm)']
plt.plot(X, linear_func(X))
plt.show()
```

#### 3. 회귀선 그래프

```python
import numpy as np

# np.polyfit: x, y, 차수를 ㅣ입력받아 최소제곱다항식에 적합
X, Y = iris['sepal length (cm)'], iris['petal length (cm)']
b1, b0 = np.polyfit(X, Y, 1)
print(b1, b0)

plt.scatter(X, Y, alpha=0.5)
plt.plot(X, b1*X + b0, color='red')
plt.show()
```

```python
import numpy as np

# np.polyfit: x, y, 차수를 ㅣ입력받아 최소제곱다항식에 적합
iris = iris.sort_values(by ='sepal length (cm)')
X, Y = iris['sepal length (cm)'], iris['petal length (cm)']
b2, b1, b0 = np.polyfit(X, Y, 2)
print(b1, b0)

plt.scatter(X, Y, alpha=0.5)
plt.plot(X, b2*X**2 + b1*X + b0, color='red')
plt.show()
```

#### 4. 꺾은선 그래프


```python
plt.plot('sepal length (cm)', 'petal length (cm)', data=iris)
plt.show()
```

```python
plt.plot('sepal length (cm)', 'petal length (cm)',
         data=iris.loc[iris['class']=='Setosa'])
plt.plot('sepal length (cm)', 'petal length (cm)',
         data=iris.loc[iris['class']=='Versicolor'])
plt.plot('sepal length (cm)', 'petal length (cm)',
         data=iris.loc[iris['class']=='Virginica'])
plt.legend(iris['class'].unique())
plt.show()
```

### 제6절. 상관관계 시각화

#### 1. 산점도 행렬

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris, alpha=0.5, figsize=(8, 8), diagonal='hist')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(iris, diag_kind='auto', hue='class')
plt.show()
```

```python
iris_corr = iris.drop(columns='class').corr(method='pearson')
iris_corr
```

```python
sns.heatmap(iris_corr, 
            xticklabels=iris_corr.columns,
            yticklabels=iris_corr.columns, 
            cmap='RdBu_r', 
            annot=True)
plt.show()
```
